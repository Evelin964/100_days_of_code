```python
import asyncio
from flight_search import FlightSearch

async def main():
    flight_search = FlightSearch()
    tequila_call_args = {
        "fly_from": 'LON',
        "fly_to": 'TYO',
        "date_from": '21/05/2024',
        "date_to": '22/05/2024',
        "curr": "EUR",
        "max_stopovers": 0
    }
    # Await the coroutine to get the actual result
    result = await flight_search.destination_search_tequila(tequila_call_args)
    flights_info = flight_search.create_flights_dict_tequila()
    print(result)
    print(flights_info) 
# Get the current event loop
loop = asyncio.get_event_loop()
# If the loop is already running, use create_task to schedule the coroutine for execution
if loop.is_running():
    task = loop.create_task(main())
else:
    # If the loop is not running, use run_until_complete (or run_forever, depending on your needs)
    loop.run_until_complete(main())
```
